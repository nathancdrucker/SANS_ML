```python
import os
```

```python
fd = '/Users/nathandrucker/Desktop/QuantumMatterGroup/Cluster/SANSML/fields'
```

```python
p = []
for file in os.listdir(fd) :
    if len(file)==12:
        a = float(file[1:3])
        b = float(file[5:7])
    elif len(file)==13:
        a = float(file[1:4])
        b = float(file[6:8]) 
    f = U.openField(os.path.join(fd,file))
    C = f.plane(z=0).topological_charge()
    p.append([b,a,C])
```

```python
fd = '/Users/nathandrucker/Desktop/QuantumMatterGroup/Cluster/SANSML/fields31_08_2020'
for file in os.listdir(fd) :
    if len(file)==12:
        a = float(file[1:3])
        b = float(file[5:7])
    elif len(file)==13:
        a = float(file[1:4])
        b = float(file[6:8])        
    f = U.openField(os.path.join(fd,file))
    C = f.plane(z=0).topological_charge()
    p.append([b,a,C])
```

```python
p=array(p)
shape(p)
```

```python
C = abs(p[:,2])
cl = []
for charge in C: 
    if charge>=10:
        cl.append(1)
    elif charge<10:
        cl.append(0)
cl = array(cl)
```

```python


fig, ax = subplots(1,1, figsize = [5,5])
t = ax.scatter(p[:,0],p[:,1], c=C, cmap='plasma',
               norm=mpl.colors.LogNorm(vmin=1e-1, vmax=C.max()))
U.colorbar(t, 'charge')
ax.set_xlabel(r'$\nu$')
ax.set_ylabel('h')
ax.set_xlim(-1,15)
ax.set_ylim(-.1,1.6)
```

```python

fig, ax = subplots(1,1, figsize = [5,5])
t = ax.scatter(p[:,0],p[:,1], c=cl, cmap='bwr')
U.colorbar(t, 'class')
ax.set_xlabel(r'$\nu$')
ax.set_ylabel('h')
ax.set_xlim(-1,15)
ax.set_ylim(-.1,1.6)
```
