```python
images = []
labels = []
shapes = []
thresh = 10 
fd = '/Users/nathandrucker/Desktop/QuantumMatterGroup/Cluster/SANSML/fields'
for file in os.listdir(fd) :   
    f = U.openField(os.path.join(fd,file))
    cross = U.crossSection(f)
    if shape(cross)[0]==50:
        C = f.plane(z=0).topological_charge()
        labels.append(int(abs(C)>=10))   
        images.append(cross.reshape(shape(cross)[0], shape(cross)[1],1))
    
fd = '/Users/nathandrucker/Desktop/QuantumMatterGroup/Cluster/SANSML/fields30_08_2020'
for file in os.listdir(fd) :
    f = U.openField(os.path.join(fd,file))
    cross = U.crossSection(f)
    if shape(cross)[0]==50:
        C = f.plane(z=0).topological_charge()
        labels.append(int(abs(C)>=10))   
        images.append(cross.reshape(shape(cross)[0], shape(cross)[1],1))
    
fd = '/Users/nathandrucker/Desktop/QuantumMatterGroup/Cluster/SANSML/fields31_08_2020'
for file in os.listdir(fd) :
    f = U.openField(os.path.join(fd,file))
    cross = U.crossSection(f)
    if shape(cross)[0]==50:
        C = f.plane(z=0).topological_charge()
        labels.append(int(abs(C)>=10))   
        images.append(cross.reshape(shape(cross)[0], shape(cross)[1],1))
   
```

```python
images = array(images).reshape(180,50,50,1)
labels = array(labels).reshape(180,1)
```

```python
import pickle


filehandler = open('./CNN/images.p', 'wb') 
pickle.dump(images, filehandler)
filehandler.close()

filehandler = open('./CNN/labels.p', 'wb') 
pickle.dump(labels, filehandler)
filehandler.close()
```
